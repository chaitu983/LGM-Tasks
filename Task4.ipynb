*   **IMPORTING THE LIBRARIES**



```python
import cv2
from PIL import Image
from IPython.display import display
from skimage import io
```



*   **LOADING THE IMAGE**



```python
img = io.imread("https://raw.githubusercontent.com/chaitu983/LGM-Tasks/main/DWARAKA.jpg")
img = cv2.resize(img, (600, 750))
display(Image.fromarray(img))
```



*   **CONVERT THE IMAGE INTO GRAY SCALE**



```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```



*   **INVERTING THE IMAGE**



```python
inverted_img = cv2.bitwise_not(gray_img)
display(Image.fromarray(inverted_img))
```



*  **BLURRING THE IMAGE USING GAUSSIAN BLUR**


```python
blurred = cv2.GaussianBlur(inverted_img, (5, 5), 0, 0)
display(Image.fromarray(blurred))
```



*   **INVERTING THE IMAGE TO GET THE FINAL SKETCH**



```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=250.0)
display(Image.fromarray(pencil_sketch))
```



*   **COMPARING THE ORIGINAL IMAGE WITH FINAL SKETCH**


```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```
